# HANDLING CATEGORICAL COLUMNS

```python
df.select_dtypes(['object']).columns
```

```python
df['Change'].value_counts()
```

```python
df['Change']=df['Change'].apply(lambda x:x.replace(',',"") )
```

```python
df['Change']=df['Change'].astype('float64')
```

```python
df['365 Day Percentage Change'].value_counts()
```

```python
df['365 Day Percentage Change']=df['365 Day Percentage Change'].apply(lambda x:float(x.replace('-','0')))
```

```python
df['365 Day Percentage Change'].dtype
```

```python
df['30 Day Percentage Change']=df['30 Day Percentage Change'].apply(lambda x:float(x.replace('-','0')))
```

```python
df['30 Day Percentage Change'].dtype
```

```python
df.shape
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
df.shape
```

# HANDLING NUMERICAL COLUMNS

```python
col_list=['Open','High','Low','Previous Close','Last Traded Price','Change','Percentage Change','Share Volume','Value (Indian Rupee)','52 Week High','52 Week Low','365 Day Percentage Change','30 Day Percentage Change']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
df[col_list]=sc.fit_transform(df[col_list])
```

```python
df.head()
```

# TRAINING AND TESTING DATA

```python
X=df.drop('Value (Indian Rupee)',axis=1)
y=df['Value (Indian Rupee)']
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=101)
```

# MODEL CREATION

```python
from sklearn.linear_model import LinearRegression
```

```python
pred=lr.predict(X_test)
```

```python
score1=lr.score(X_train,y_train)
```
